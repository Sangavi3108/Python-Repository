```python
days=[1,2,3,4,5,6,7] # weekdays
max_t=[50,51,52,48,47,49,46] # max temp recorder at that day
min_t=[43,42,40,44,33,35,37] # min temp
avg_t=[45,48,48,46,40,42,41] # avg temp
```

```python
# multiple plots in a same graph
plt.plot(days , max_t )
plt.plot(days , min_t )
plt.plot(days , avg_t )

# AXIS NAME
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('Weather')
```

```python
#legend

plt.plot(days , max_t, label = 'max')
plt.plot(days , min_t, label = 'min' )
plt.plot(days , avg_t, label = 'avg' )

# AXIS NAME
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('Weather')
plt.legend()
```

```python
#legend

plt.plot(days , max_t, label = 'max')
plt.plot(days , min_t, label = 'min' )
plt.plot(days , avg_t, label = 'avg' )

# AXIS NAME
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('Weather')
plt.legend(loc='right')
```

```python
# shadow

plt.plot(days , max_t, label = 'max')
plt.plot(days , min_t, label = 'min' )
plt.plot(days , avg_t, label = 'avg' )

# AXIS NAME
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('Weather')
plt.legend(loc='right' , shadow=True)
```

```python
# fontsize
plt.plot(days , max_t, label = 'max')
plt.plot(days , min_t, label = 'min' )
plt.plot(days , avg_t, label = 'avg' )

# AXIS NAME
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('Weather')
plt.legend(loc='right' , shadow=True , fontsize='small')
```

```python
# grid

plt.plot(days , max_t, label = 'max')
plt.plot(days , min_t, label = 'min' )
plt.plot(days , avg_t, label = 'avg' )

# AXIS NAME
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('Weather')
plt.legend(loc='right' , shadow=True , fontsize='small')
plt.grid()
```

```python
# bar graph

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
expenditure = [80,120,70,44]
```

```python
plt.bar(company , revenue , label = 'Revenue' , width = 0.4)
plt.bar(company , expenditure , label = 'exp' , width = 0.4)

plt.xlabel('company')
plt.ylabel('revenue(bln)')
plt.title('Us Technologogy stocks')
plt.legend()
plt.grid()
```

```python
import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2 , revenue , label = 'Revenue' , width = 0.4)
plt.bar(xpos+0.2 , expenditure , label = 'exp' , width = 0.4)

plt.yticks([10,20,30,40,50])
plt.ylabel('revenue')
plt.title('Us Technologogy stocks')
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(4, 10))
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
