# Merge

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
pd.merge(df1 , df2 , on='city')
```

```python
# types of joins

# 1 = left join
# 2 = right join
# 3 = inner join
# 4 = outer join
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1 , df2 , on='city')
```

```python
# left join
pd.merge(df2 , df1 , on='city' , how = 'left')
```

```python
pd.merge(df1 , df2 , on='city' , how = 'right')
```

```python
pd.merge(df1 , df2 , on='city' , how='inner')
```

```python
pd.merge(df1 , df2 , on='city' , how='outer')
```

```python
#indicator flag
df3 = pd.merge(df1 , df2 , on='city' , how='outer' , indicator=True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1 , df2 , on='city' , how='outer' , indicator=True , suffixes = ('_first' , '_second'))
df3
```

# Pivot

```python
df = pd.read_csv("dataset/weather.csv")
df
```

```python
#Q1  suppose you want to transform thi data where the index is date and columns are city
df.pivot(index='date' , columns='city')
```

```python
df = pd.read_csv('dataset/weather2.csv')
df
```

```python
# Q1 - you want the avrage of the temp and humidity on a particular day

df.pivot_table(index='city' , columns='date' , aggfunc='median')
```

```python
# stack and unstack
df = pd.read_csv('datasets/avocado.csv')
df
```

```python
df_pivot = df.pivot(index = 'Region' , columns='Type')
df_pivot
```

```python
# stack - stack will convert the columns into rows
B = df_pivot.stack(level = 1)
B
```

```python
# unstack - stack will convert the rows into columns
B.unstack(level = 0)
```
