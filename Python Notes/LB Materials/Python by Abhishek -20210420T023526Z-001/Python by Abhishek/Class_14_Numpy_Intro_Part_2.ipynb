```python
import numpy as np
```

```python
arr = np.arange(25, 55).reshape(5, 6)
arr
```

```python
arr > 40
```

```python
arr[arr>40]
```

```python
# Finding the total number of elements
np.sum(arr>=25)
```

```python
# Find the total no of elements in every row in the array where the element is greater than 40
np.sum(arr>40, axis = 0)
```

```python
np.sum(arr>40, axis = 1)
```

```python
np.all(arr>40)
```

```python
np.any(arr>40)
```

```python
# Iterating over an array
arr1 = np.array([10,20,30,40,50,60,70])
for i in arr1:
    print(i)


```

```python
for i in arr:
    print(i)
```

```python
for i in arr:
    for j in i:
        print(j)
```

```python
for i in np.nditer(arr, order = 'C'):    # Row major form
    print(i)
```

```python
for i in np.nditer(arr, order = 'F'):    # Col major form
    print(i)
```

```python
# Split an array

arr = np.arange(25, 61).reshape(6, 6)
arr
```

```python
# Splitting an array in two equal array along the rows - Default split is done along the rows
np.split(arr, 2, axis = 0)
```

```python
# Splitting an array in two equal array along the columns
np.split(arr, 2, axis = 1)
```

```python
np.split(arr, 6, axis = 1)
```

```python
np.split(arr, 6, axis = 0)
```

```python
np.split(arr, [1, 3, 5], axis = 1)
```

# Concatenate

```python
a = np.array([["New York","Maryland"],["Albany","Annapolis"]])

b = np.array([["Ohio","Arizona"],["Columbus","Phoenix"]])
```

```python
a
```

```python
b
```

```python
np.concatenate((a,b), axis = 0)   # Along the rows then no of column should match
```

```python
np.concatenate((a,b), axis = 1)   # Along the cols then no of rows should match
```

# Image Manipulation

```python
from skimage import data
import matplotlib.pyplot as plt
```

```python
astronaut = data.astronaut()
plt.imshow(astronaut)
```

```python
astronaut
```

```python
astronaut.shape
```

```python
astronaut.size
```

```python
# Cropping an image - using slicing

plt.imshow(astronaut[100:, 50:])

```

```python
h1, h2, h3, h4 = np.split(astronaut, 4, axis = 0)
plt.imshow(h1)
```

```python
plt.imshow(h2)
```

```python
plt.imshow(h3)
```

```python
plt.imshow(h4)
```

```python
plt.imshow(np.concatenate((h2,h4,h3,h1)))
```

```python
v1, v2 = np.split(astronaut, 2, axis = 1)
plt.imshow(v1)
```

```python
plt.imshow(v1)
```

```python
plt.imshow(np.concatenate((v1,v2)))
```
